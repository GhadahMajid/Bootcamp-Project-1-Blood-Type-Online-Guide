```python

def compatibility(BT):
    BTC = {"Donor" : {
        "A": ["A","AB"],
        "B": ["B","AB"],
        "O": ["A","B","O","AB"],
        "AB": ["AB"]
    }, "Recipient" : {
        "A": ["O","A"],
        "B": ["O","B"],
        "O": ["O"],
        "AB": ["A","B","O","AB"]
    }
                }
    return BTC["Donor"][BT] , BTC["Recipient"][BT]

    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
    
def Genetically(BT):
    BTG = {"AA": ["A", "O"],
           "AB": ["A","B","O","AB"],
           "AAB": ["A","B","AB"],
           "AO": ["A","O"],
           "BB": ["B","O"],
           "BAB": ["A","B","AB"],
           "BO": ["B","O"],
           "ABAB": ["A","B","AB"],
           "ABO": ["A","B"],
           "OO": ["O"],
          }
    return BTG[BT]
    
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #

BloodType = ["A","B","O","AB"]
MenuLoop = True
print('\033[1m' + 'Welcome to Blood Type Online Guide \U0001F489 ..' + '\033[0m')

while(MenuLoop):
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
    menu = input("Enter your choice number:\n \
    1. Check your blood type Donor-Recipient compatibility. \n \
    2. Check your blood type genetically.\n")
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
    if menu == '1':
        print("1")
        BT = input("Enter your Blood Type \U0001FA78: ").upper()
        if BT in BloodType:
            BTPercent = lambda BT: 16 if BT == 'B' else 45 if BT == 'O' else 33.50 \
            if BT == 'A' else 5.50 if BT == 'AB' else ' '
            print("You are one of the",'\033[1m' + str(BTPercent(BT)) + '%\033[0m',\
                  "of the world who has",'\033[1m' + BT + '\033[0m',"blood type \U0001F30D.")
            result = compatibility(BT)
            print("You can donate your blood to: ",'\033[1m' + str(result[0]) + '\033[0m')
            print("You can receive blood from: ",'\033[1m' + str(result[1]) + '\033[0m')
        else :
            print("Invalid Blood Type. It must be A, B, O, or AB only.")
            continue
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
    elif menu == '2':
        print("2")
        FBT = input("Enter Father Blood Type \U0001FA78: ").upper()
        MBT = input("Enter Mother Blood Type \U0001FA78: ").upper()
        if FBT in BloodType and MBT in BloodType:
            result = Genetically(''.join(sorted(FBT+MBT)))
            print("Children's possible blood types from", '\033[1m' + FBT + '\033[0m',"and",\
                  '\033[1m' + MBT + '\033[0m', "parents is:",'\033[1m' + str(result) + '\033[0m')
        else :
            print("Invalid blood type. It must be A, B, O, or AB only.")
            continue
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
    else :
        print("Please choose number from the menu.")
    # - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - #
    cont = input("Continue (y/n)?")
    if cont == 'n':
        MenuLoop = False
    print("- - - - - - - - - - - - - - - - - - - - - - - - - - - - - -")
```
